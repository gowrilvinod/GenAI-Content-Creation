```python
from transformers import pipeline

# Initialize the GPT-2 model pipeline from Hugging Face
gpt2_pipeline = pipeline('text-generation', model="gpt2")

# Define the conversation prompts
system_prompt = "You are Della, an AI model designed to assist with various tasks."
ai_prompt = "How can I assist you today?"
human_prompt = "Who is the father of the nation in India?"

# Function to generate AI responses using GPT-2 pipeline
def generate_response(prompt, max_new_tokens=50):
    # Generate response
    response = gpt2_pipeline(prompt, max_new_tokens=max_new_tokens, truncation=True)[0]['generated_text']
    # Clean response
    return response.strip()

# Start LangChain by generating AI response to the system prompt
ai_response_1 = generate_response(system_prompt + "\n" + ai_prompt)

# Generate AI response to the human prompt using the last AI response as input
ai_response_2 = generate_response(ai_response_1 + "\n" + human_prompt)

# Print out the conversation for clarity
print(f"System: {system_prompt}")
print(f"AI 1: {ai_response_1}")
print(f"Human: {human_prompt}")
print(f"AI 2: {ai_response_2}")

# Return AI's response to the human prompt
print(f"Final AI response: {ai_response_2}")
```
